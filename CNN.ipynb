```python
"""
PyTorch CNN+LSTM Emotion Recognition (audio MFCC)  optimized version
- Keeps your directory layout
- Batch size = 64, Epochs = 50
- Mixed precision (AMP) + CUDA optimizations for RTX 5060
- Shows before/after MFCC, accuracy vs epoch, confusion matrix, multi-class ROC
- Saves best model and emotions.npy
"""

import os
import random
import copy
from glob import glob
from typing import List

import numpy as np
import librosa
import matplotlib.pyplot as plt
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.metrics import confusion_matrix, roc_curve, auc, accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score


# ---------------------------
# Config (keeps your directories & filenames)
# ---------------------------
DATA_DIRS = [
    r"K:\Code\Project\Research Paper\Emotion Detection\voice_data\train_data\SER",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_01",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_02",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_03",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_04",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_05",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_06",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_07",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_08",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_09",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_10",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_11",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_12",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_13",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_14",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_15",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_16",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_17",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_18",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_19",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_20",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_21",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_22",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_23",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive\audio_speech_actors_01-24\Actor_24",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive2\Angry",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive2\Happy",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive2\Natural",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive2\Sad",
    r"K:\Code\Project\Research Paper\Emotion Detection\archive2\Surprised"
]

MODEL_PATH = r"K:\Code\Project\Research Paper\Emotion Detection\Code\cnn_emotion_model_speech_final.pth"
EMOTIONS_NPY = r"K:\Code\Project\Research Paper\Emotion Detection\Code\emotions.npy"
TEST_FILE = r"K:\Code\Project\Research Paper\Emotion Detection\voice_data\test_data\self.wav"  # optional

# Audio / MFCC params
SR = 22050
N_MFCC = 40
MAX_PAD_LEN = 174  # keep as in your TF code so shapes match

# Training hyperparams
BATCH_SIZE = 64
EPOCHS = 50
LR = 1e-3

# Device / GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f" Device: {device} (CUDA available: {torch.cuda.is_available()})")

# Performance options
NUM_WORKERS = 0
PIN_MEMORY = True if torch.cuda.is_available() else False

# Reproducibility
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# ---------------------------
# Feature extraction (MFCC)
# ---------------------------
def extract_mfcc(file_path, max_pad_len=MAX_PAD_LEN, sr=SR, n_mfcc=N_MFCC):
    y, _ = librosa.load(file_path, sr=sr)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    # pad or truncate along time axis
    if mfcc.shape[1] < max_pad_len:
        pad_width = max_pad_len - mfcc.shape[1]
        mfcc = np.pad(mfcc, pad_width=((0,0),(0,pad_width)), mode='constant')
    else:
        mfcc = mfcc[:, :max_pad_len]
    return mfcc.astype(np.float32)

# ---------------------------
# Load dataset paths & labels (detect labels from filenames like before)
# ---------------------------
def scan_and_build_list(dirs: List[str]):
    X_paths = []
    y_labels = []
    emotions = []
    for d in dirs:
        if not os.path.exists(d):
            print(f" directory not found, skipping: {d}")
            continue
        for fname in os.listdir(d):
            if not fname.lower().endswith('.wav'):
                continue
            # your original TF code parsed label as last underscore part
            label = fname.split("_")[-1].replace(".wav", "")
            if label not in emotions:
                emotions.append(label)
            path = os.path.join(d, fname)
            X_paths.append(path)
            y_labels.append(emotions.index(label))  # index is created on-the-fly
    # Note: the way we assign indices above depends on order discovered; to be safer we will re-map:
    emotions = sorted(list(set(emotions)))
    label_map = {emo: idx for idx, emo in enumerate(emotions)}
    y_labels = [label_map[fname.split("_")[-1].replace(".wav", "")] for fname in [os.path.basename(p) for p in X_paths]]
    return X_paths, y_labels, emotions

# ---------------------------
# Dataset class (returns MFCC tensors)
# ---------------------------
class MFCCDataset(Dataset):
    def __init__(self, file_paths, labels):
        self.paths = file_paths
        self.labels = labels

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        p = self.paths[idx]
        mfcc = extract_mfcc(p)  # shape (n_mfcc, max_pad_len)
        # normalize per-sample (zero mean, unit var)  common for MFCCs
        mfcc = (mfcc - np.mean(mfcc)) / (np.std(mfcc) + 1e-9)
        # add channel dim -> (1, n_mfcc, max_pad_len)
        mfcc = np.expand_dims(mfcc, axis=0)
        return torch.from_numpy(mfcc), torch.tensor(self.labels[idx], dtype=torch.long)

# ---------------------------
# Model: simple CNN -> LSTM -> FC
# ---------------------------
class CNN_LSTM(nn.Module):
    def __init__(self, n_classes):
        super().__init__()
        # convolutional front-end (small and efficient)
        self.conv = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1),  # (B,32,40,174)
            nn.ReLU(),
            nn.BatchNorm2d(32),
            nn.MaxPool2d((2,2)),                         # -> (B,32,20,87)
            nn.Dropout(0.2),

            nn.Conv2d(32, 64, kernel_size=3, padding=1), # -> (B,64,20,87)
            nn.ReLU(),
            nn.BatchNorm2d(64),
            nn.MaxPool2d((2,2)),                         # -> (B,64,10,43)
            nn.Dropout(0.2),
        )
        # We'll treat the spatial width dimension (time) as sequence length for LSTM:
        # After conv -> (B, C, Freq, TimeSteps). We'll reshape to (B, TimeSteps, C*Freq) for LSTM.
        self.lstm_input_size = 64 * 10  # channels * freq_bins after pooling (freq_bins ~10)
        self.lstm = nn.LSTM(input_size=self.lstm_input_size, hidden_size=256, num_layers=2,
                            batch_first=True, dropout=0.25)
        self.classifier = nn.Sequential(
            nn.Dropout(0.3),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, n_classes)
        )

    def forward(self, x):
        # x: (B,1,40,174)
        b = x.size(0)
        c = self.conv(x)  # (B,64,10,T)
        # ensure correct shape
        # c shape: (B, C, Freq, Time)
        B, C, Freq, Time = c.shape
        # reshape to sequence: (B, Time, C*Freq)
        seq = c.permute(0, 3, 1, 2).contiguous().view(B, Time, C * Freq)
        lstm_out, _ = self.lstm(seq)  # (B, Time, hidden)
        last = lstm_out[:, -1, :]     # (B, hidden)
        out = self.classifier(last)   # (B, n_classes)
        return out

# ---------------------------
# Plotting helpers
# ---------------------------
# ---------------------------
# Plotting helpers (waveform  MFCC)
# ---------------------------
def plot_waveform_and_mfcc(file_path):
    """Show raw waveform vs MFCC after preprocessing."""
    y, sr = librosa.load(file_path, sr=SR)
    mfcc = extract_mfcc(file_path)
    mfcc_pre = (mfcc - np.mean(mfcc)) / (np.std(mfcc) + 1e-9)

    fig, axs = plt.subplots(1, 2, figsize=(12, 4))

    # Left: Raw waveform
    axs[0].plot(np.linspace(0, len(y)/sr, len(y)), y, color='steelblue')
    axs[0].set_title("Raw Audio Waveform")
    axs[0].set_xlabel("Time (s)")
    axs[0].set_ylabel("Amplitude")

    # Right: MFCC
    img = axs[1].imshow(mfcc_pre, origin='lower', aspect='auto', cmap='viridis')
    axs[1].set_title("MFCC After Preprocessing")
    axs[1].set_xlabel("Time Frames")
    axs[1].set_ylabel("MFCC Coefficients")
    fig.colorbar(img, ax=axs[1], orientation='vertical', fraction=0.046)

    plt.tight_layout()
    plt.show()


def plot_accuracy_curve(train_accs, val_accs):
    epochs = range(1, len(train_accs)+1)
    plt.figure(figsize=(7,4))
    plt.plot(epochs, train_accs, marker='o', label='Train Acc')
    plt.plot(epochs, val_accs, marker='o', label='Val Acc')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy (%)')
    plt.title('Accuracy vs Epoch')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_confusion_matrix_and_roc(y_true, y_pred_probs, emotions):
    y_pred = np.argmax(y_pred_probs, axis=1)
    labels_idx = list(range(len(emotions)))
    cm = confusion_matrix(y_true, y_pred, labels=labels_idx)
    # Confusion matrix
    fig, ax = plt.subplots(figsize=(6,6))
    im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    ax.figure.colorbar(im, ax=ax)
    ax.set(xticks=np.arange(len(emotions)), yticks=np.arange(len(emotions)),
           xticklabels=emotions, yticklabels=emotions,
           ylabel='True label', xlabel='Predicted label', title='Confusion Matrix')
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right")
    # annotate
    thresh = cm.max()/2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i,j], 'd'),
                    ha="center", va="center",
                    color="white" if cm[i,j] > thresh else "black")
    plt.tight_layout()
    plt.show()

    # ROC: binarize and compute per-class ROC
    try:
        y_true_bin = label_binarize(y_true, classes=labels_idx)
        n_classes = y_true_bin.shape[1]
        plt.figure(figsize=(8,6))
        for i in range(n_classes):
            fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_pred_probs[:, i])
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=f"{emotions[i]} (AUC={roc_auc:.2f})")
        # micro-average
        fpr, tpr, _ = roc_curve(y_true_bin.ravel(), y_pred_probs.ravel())
        plt.plot(fpr, tpr, label=f"micro-average (AUC={auc(fpr, tpr):.2f})", linestyle='--', linewidth=2)
        plt.plot([0,1],[0,1],'k--', linewidth=0.6)
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Multi-class ROC')
        plt.legend(fontsize='small', loc='lower right')
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(" ROC plotting failed (maybe missing classes in y_true):", e)

# ---------------------------
# Training & evaluation
# ---------------------------
def train_and_evaluate(model, train_loader, val_loader, optimizer, criterion, epochs, model_path):
    scaler = torch.cuda.amp.GradScaler() if torch.cuda.is_available() else None
    best_val_acc = -1.0
    best_state = None
    train_accs = []
    val_accs = []

    for epoch in range(epochs):
        model.train()
        running_correct = 0
        running_total = 0
        running_loss = 0.0

        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}", leave=False)
        for xb, yb in pbar:
            xb = xb.to(device, non_blocking=True)
            yb = yb.to(device, non_blocking=True)

            optimizer.zero_grad()
            if scaler:
                with torch.cuda.amp.autocast():
                    out = model(xb)
                    loss = criterion(out, yb)
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()
            else:
                out = model(xb)
                loss = criterion(out, yb)
                loss.backward()
                optimizer.step()

            preds = out.argmax(dim=1)
            running_total += yb.size(0)
            running_correct += (preds == yb).sum().item()
            running_loss += loss.item() * xb.size(0)

            train_acc = 100.0 * running_correct / running_total
            avg_loss = running_loss / running_total
            pbar.set_postfix({"Loss": f"{avg_loss:.4f}", "TrainAcc": f"{train_acc:.2f}%"})

        epoch_train_acc = 100.0 * running_correct / running_total
        train_accs.append(epoch_train_acc)

        # validation
        model.eval()
        val_preds = []
        val_probs = []
        val_targets = []
        val_loss = 0.0
        val_total = 0
        with torch.no_grad():
            for xb, yb in val_loader:
                xb = xb.to(device, non_blocking=True)
                yb = yb.to(device, non_blocking=True)
                out = model(xb)
                probs = F.softmax(out, dim=1)
                loss = criterion(out, yb)
                val_loss += loss.item() * xb.size(0)
                val_total += xb.size(0)
                val_preds.append(probs.argmax(dim=1).cpu().numpy())
                val_probs.append(probs.cpu().numpy())
                val_targets.append(yb.cpu().numpy())

        val_preds = np.concatenate(val_preds) if val_preds else np.array([])
        val_probs = np.vstack(val_probs) if val_probs else np.array([])
        val_targets = np.concatenate(val_targets) if val_targets else np.array([])

        epoch_val_acc = 100.0 * (val_preds == val_targets).sum() / len(val_targets) if len(val_targets)>0 else 0.0
        val_accs.append(epoch_val_acc)

        avg_val_loss = val_loss / val_total if val_total>0 else 0.0
        print(f"Epoch [{epoch+1}/{epochs}] Train Acc: {epoch_train_acc:.2f}% | Val Acc: {epoch_val_acc:.2f}% | Val Loss: {avg_val_loss:.4f}")

        # Save best
        if epoch_val_acc > best_val_acc:
            best_val_acc = epoch_val_acc
            best_state = {
                'epoch': epoch+1,
                'model_state': copy.deepcopy(model.state_dict()),
                'optimizer_state': optimizer.state_dict(),
                'val_acc': best_val_acc
            }
            torch.save(best_state['model_state'], model_path)
            print(f" Best model saved (val acc {best_val_acc:.2f}%) to {model_path}")

    return {
        'train_accs': train_accs,
        'val_accs': val_accs,
        'best_state': best_state,
        'val_preds': val_preds,
        'val_probs': val_probs,
        'val_targets': val_targets
    }

# ---------------------------
# Main routine
# ---------------------------
def main():
    # Step 1: scan dataset paths & labels
    print(" Scanning audio directories and building file list...")
    X_paths, y_labels, emotions = scan_and_build_list(DATA_DIRS)
    if len(X_paths) == 0:
        raise RuntimeError("No WAV files found. Check DATA_DIRS paths.")
    print(f" Found {len(X_paths)} audio files across {len(emotions)} emotions: {emotions}")

    # Save emotions mapping
    os.makedirs(os.path.dirname(EMOTIONS_NPY), exist_ok=True)
    np.save(EMOTIONS_NPY, emotions)

    # Train/val split (stratify to keep classes balanced)
    stratify = y_labels if len(set(y_labels)) > 1 else None
    X_train, X_val, y_train, y_val = train_test_split(X_paths, y_labels, test_size=0.15, random_state=SEED, stratify=stratify)

    train_ds = MFCCDataset(X_train, y_train)
    val_ds = MFCCDataset(X_val, y_val)

    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=PIN_MEMORY)
    val_loader = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=PIN_MEMORY)

    print(" Showing raw audio waveform and MFCC after preprocessing...")
    plot_waveform_and_mfcc(X_train[0])


    # Build model
    n_classes = len(emotions)
    model = CNN_LSTM(n_classes).to(device)
    print(f"[INFO] Model built with {n_classes} output classes.")

    # Option: load or train
    choice = input("Do you want to load the saved model? (yes/no): ").strip().lower()
    if choice == "yes" and os.path.exists(MODEL_PATH):
        print(" Loading saved model from", MODEL_PATH)
        state = torch.load(MODEL_PATH, map_location=device)
        model.load_state_dict(state)
        model.to(device)
        # Evaluate on validation set and show plots
        model.eval()
        all_probs = []
        all_targets = []
        with torch.no_grad():
            for xb, yb in val_loader:
                xb = xb.to(device)
                out = model(xb)
                probs = F.softmax(out, dim=1).cpu().numpy()
                all_probs.append(probs)
                all_targets.append(yb.numpy())
        if all_probs:
            y_score = np.vstack(all_probs)
            y_true = np.concatenate(all_targets)
            plot_confusion_matrix_and_roc(y_true, y_score, emotions)
        # single test predict
        if os.path.exists(TEST_FILE):
            mfcc = extract_mfcc(TEST_FILE)
            mfcc = (mfcc - np.mean(mfcc)) / (np.std(mfcc) + 1e-9)
            x = torch.from_numpy(np.expand_dims(mfcc, axis=(0,1))).to(device)
            with torch.no_grad():
                out = model(x)
                p = F.softmax(out, dim=1).cpu().numpy()[0]
                pred = emotions[int(p.argmax())]
                print(f"[PREDICT] {TEST_FILE} -> {pred} (conf {p.max():.3f})")
        return

    # Train path
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=LR)

    stats = train_and_evaluate(model, train_loader, val_loader, optimizer, criterion, epochs=EPOCHS, model_path=MODEL_PATH)

    # Plot accuracy vs epoch
    plot_accuracy_curve(stats['train_accs'], stats['val_accs'])

    # Load best model for final evaluation (if saved)
    if stats['best_state'] is not None:
        model.load_state_dict(stats['best_state']['model_state'])
        print(f"[INFO] Loaded best model (epoch {stats['best_state']['epoch']}, val acc {stats['best_state']['val_acc']:.2f}%)")

    # Evaluate and plot confusion matrix & ROC with validation set outputs
    if stats['val_probs'] is not None and len(stats['val_probs'])>0:
        plot_confusion_matrix_and_roc(stats['val_targets'], stats['val_probs'], emotions)
    else:
        print(" No validation predictions to plot.")

    # Final single-file prediction (optional)
    if os.path.exists(TEST_FILE):
        mfcc = extract_mfcc(TEST_FILE)
        mfcc = (mfcc - np.mean(mfcc)) / (np.std(mfcc) + 1e-9)
        x = torch.from_numpy(np.expand_dims(mfcc, axis=(0,1))).to(device)
        with torch.no_grad():
            out = model(x)
            p = F.softmax(out, dim=1).cpu().numpy()[0]
            pred = emotions[int(p.argmax())]
            print(f"[PREDICT] {TEST_FILE} -> {pred} (conf {p.max():.3f})")
        # ----- Final evaluation metrics -----
    if stats['val_probs'] is not None and len(stats['val_probs']) > 0:
        y_true = np.array(stats['val_targets'])
        y_pred = np.argmax(stats['val_probs'], axis=1)

        precision = precision_score(y_true, y_pred, average='macro', zero_division=0)
        recall = recall_score(y_true, y_pred, average='macro', zero_division=0)
        f1 = f1_score(y_true, y_pred, average='macro', zero_division=0)

        precision_w = precision_score(y_true, y_pred, average='weighted', zero_division=0)
        recall_w = recall_score(y_true, y_pred, average='weighted', zero_division=0)
        f1_w = f1_score(y_true, y_pred, average='weighted', zero_division=0)

        print("\n===== Evaluation Metrics =====")
        print(f" Macro Precision: {precision:.4f}")
        print(f" Macro Recall:    {recall:.4f}")
        print(f" Macro F1-Score:  {f1:.4f}")
        print("---------------------------------")
        print(f" Weighted Precision: {precision_w:.4f}")
        print(f" Weighted Recall:    {recall_w:.4f}")
        print(f" Weighted F1-Score:  {f1_w:.4f}")
        print("=================================\n")

    print("[DONE]")

if __name__ == "__main__":
    main()
```

```python
# optimized_gpu_training.py
import os
import random
import copy
import numpy as np
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, roc_curve, auc, classification_report
from sklearn.preprocessing import label_binarize

# -----------------------
# Config / Paths
# -----------------------
ROOT_DATA_DIRS = [
    r"K:\Code\Project\Research Paper\Emotion Detection\facial_data\archive"
]

DEFAULT_MODEL_PATH = r"K:\Code\Project\Research Paper\Emotion Detection\cnn_lstm_face_emotion.pth"
EMOTIONS_NPY = r"K:\Code\Project\Research Paper\Emotion Detection\emotions_face.npy"
TEST_IMAGE = r"K:\Code\Project\Research Paper\Emotion Detection\test5.jpg"  # example test path

# Hyperparams
SEED = 42
BATCH_SIZE = 64
LR = 1e-4
EPOCHS = 40
IMG_SIZE = 224
NUM_WORKERS = 0  # per user request

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"[INFO] Using device: {device}")

# cudnn and precision hints
torch.backends.cudnn.benchmark = True
torch.backends.cudnn.enabled = True
# optional precision hint (PyTorch >= 2.0)
try:
    torch.set_float32_matmul_precision("high")
except Exception:
    pass

# Reproducibility
def set_seed(seed=SEED):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed()

# -----------------------
# Dataset
# -----------------------
class FaceEmotionDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        p = self.image_paths[idx]
        label = self.labels[idx]
        img = Image.open(p).convert("RGB")
        if self.transform:
            img_t = self.transform(img)
        else:
            img_t = transforms.ToTensor()(img)
        # return tensor image and integer label
        return img_t, int(label)

# -----------------------
# Model 
# -----------------------
class CNN_LSTM_Emotion(nn.Module):
    def __init__(self, num_emotions):
        super().__init__()
        # ResNet18 pretrained backbone (remove final pooling & fc)
        backbone = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
        # Keep layers up to last conv block (exclude avgpool & fc)
        self.cnn = nn.Sequential(*list(backbone.children())[:-2])  # -> (B, 512, H', W')
        conv_out_channels = 512
        self.conv_out_channels = conv_out_channels

        # LSTM on flattened spatial locations
        self.lstm = nn.LSTM(
            input_size=conv_out_channels,
            hidden_size=512,
            num_layers=2,
            batch_first=True,
            dropout=0.25,
            bidirectional=False,
        )

        self.fc1 = nn.Linear(512, 128)
        self.dropout = nn.Dropout(0.25)
        self.fc_out = nn.Linear(128, num_emotions)

    def forward(self, x):
        b = x.size(0)
        feat = self.cnn(x)  # (B, 512, h, w)
        feat = feat.view(b, self.conv_out_channels, -1)  # (B, 512, S)
        feat = feat.permute(0, 2, 1)  # (B, S, 512)
        lstm_out, _ = self.lstm(feat)  # (B, S, hidden)
        last = lstm_out[:, -1, :]  # (B, hidden)
        x = F.relu(self.fc1(last))
        x = self.dropout(x)
        x = self.fc_out(x)
        return x

# -----------------------
# Utilities
# -----------------------
def load_multiple_datasets(dir_list):
    image_paths, labels = [], []
    emotions = set()
    for root_dir in dir_list:
        print(f"Scanning {root_dir} ...")
        for subdir, _, files in os.walk(root_dir):
            if not files:
                continue
            img_files = [f for f in files if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
            if not img_files:
                continue
            emotion = os.path.basename(subdir).lower()
            emotions.add(emotion)
            for fname in img_files:
                image_paths.append(os.path.join(subdir, fname))
                labels.append(emotion)
    emotions = sorted(list(emotions))
    emotion_to_idx = {emo: i for i, emo in enumerate(emotions)}
    numeric_labels = [emotion_to_idx[l] for l in labels]
    print(f"Found {len(image_paths)} images across {len(emotions)} emotions: {emotions}")
    return image_paths, numeric_labels, emotions

# Plot helpers (unchanged)
def show_before_after(sample_path, preprocess_transform, unnormalize):
    img = Image.open(sample_path).convert("RGB")
    fig, axs = plt.subplots(1, 2, figsize=(9, 4))
    axs[0].imshow(img)
    axs[0].set_title("Original")
    axs[0].axis("off")
    x = preprocess_transform(img).cpu()
    x_vis = unnormalize(x)
    x_vis = x_vis.permute(1, 2, 0).numpy()
    x_vis = np.clip(x_vis, 0, 1)
    axs[1].imshow(x_vis)
    axs[1].set_title("After Preprocessing")
    axs[1].axis("off")
    plt.tight_layout()
    plt.show()

def plot_accuracy(train_accs, val_accs):
    epochs = np.arange(1, len(train_accs) + 1)
    plt.figure(figsize=(7,4))
    plt.plot(epochs, train_accs, marker='o', label='Train Acc')
    plt.plot(epochs, val_accs,   marker='o', label='Val Acc')
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy (%)")
    plt.title("Training and Validation Accuracy")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(cm, labels):
    fig, ax = plt.subplots(figsize=(6,6))
    im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    ax.figure.colorbar(im, ax=ax)
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           xticklabels=labels, yticklabels=labels,
           ylabel='True label',
           xlabel='Predicted label',
           title='Confusion Matrix')
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], 'd'),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.show()

def plot_multiclass_roc(y_true, y_score, n_classes, labels):
    plt.figure(figsize=(8,6))
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_true[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
        plt.plot(fpr[i], tpr[i], lw=1.5, label=f'{labels[i]} (AUC = {roc_auc[i]:.2f})')
    fpr["micro"], tpr["micro"], _ = roc_curve(y_true.ravel(), y_score.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
    plt.plot(fpr["micro"], tpr["micro"], label=f'micro (AUC = {roc_auc["micro"]:.2f})', linestyle='--', linewidth=2)
    plt.plot([0, 1], [0, 1], 'k--', lw=0.6)
    plt.xlim([-0.01, 1.01])
    plt.ylim([-0.01, 1.01])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Multi-class ROC')
    plt.legend(loc='lower right', fontsize='small')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# -----------------------
# Training loop with mixed precision + optional torch.compile
# -----------------------
from torch.cuda.amp import autocast, GradScaler

def train_model(model, train_loader, val_loader, criterion, optimizer, epochs, start_epoch=0, scheduler=None):
    model.to(device)
    
    scaler = GradScaler()
    best_val_acc = -1.0
    best_state = None
    train_acc_history, val_acc_history = [], []

    for epoch in range(start_epoch, epochs):
        model.train()
        running_loss = 0.0
        running_correct = 0
        running_total = 0

        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}", leave=False)
        for imgs, labels in pbar:
            # ensure tensors and async transfer
            imgs = imgs.to(device, non_blocking=True)
            labels = labels.long().to(device, non_blocking=True)

            optimizer.zero_grad()
            # mixed precision forward/backward
            with autocast():
                outputs = model(imgs)
                loss = criterion(outputs, labels)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            running_loss += loss.item() * imgs.size(0)
            preds = outputs.argmax(dim=1)
            running_total += labels.size(0)
            running_correct += (preds == labels).sum().item()

            train_acc = 100.0 * running_correct / running_total
            avg_loss = running_loss / running_total
            pbar.set_postfix({
                "Loss": f"{avg_loss:.4f}",
                "TrainAcc": f"{train_acc:.2f}%"
            })

        epoch_train_acc = 100.0 * running_correct / running_total
        train_acc_history.append(epoch_train_acc)

        # Validation
        model.eval()
        val_correct = 0
        val_total = 0
        val_loss = 0.0
        all_targets = []
        all_probs = []
        with torch.no_grad():
            for imgs, labels in val_loader:
                imgs = imgs.to(device, non_blocking=True)
                labels = labels.long().to(device, non_blocking=True)
                with autocast():
                    outputs = model(imgs)
                    loss = criterion(outputs, labels)
                val_loss += loss.item() * imgs.size(0)
                probs = F.softmax(outputs, dim=1)
                preds = probs.argmax(dim=1)
                val_total += labels.size(0)
                val_correct += (preds == labels).sum().item()
                all_targets.append(labels.cpu().numpy())
                all_probs.append(probs.cpu().numpy())

        epoch_val_acc = 100.0 * val_correct / val_total if val_total > 0 else 0.0
        val_acc_history.append(epoch_val_acc)
        avg_val_loss = val_loss / val_total if val_total > 0 else 0.0

        if scheduler:
            # scheduler might expect metric or epoch; using val loss (ReduceLROnPlateau expects metric)
            try:
                if isinstance(scheduler, torch.optim.lr_scheduler.ReduceLROnPlateau):
                    scheduler.step(avg_val_loss)
                else:
                    scheduler.step()
            except Exception:
                pass

        print(f"Epoch [{epoch+1}/{epochs}] Train Acc: {epoch_train_acc:.2f}% | Val Acc: {epoch_val_acc:.2f}% | Val Loss: {avg_val_loss:.4f}")

        # Save best
        if epoch_val_acc > best_val_acc:
            best_val_acc = epoch_val_acc
            best_state = {
                'epoch': epoch+1,
                'model_state': copy.deepcopy(model.state_dict()),
                'optimizer_state': optimizer.state_dict(),
                'val_acc': best_val_acc
            }
            torch.save(best_state['model_state'], DEFAULT_MODEL_PATH)
            print(f"Best model saved with Val Acc: {best_val_acc:.2f}% at epoch {epoch+1}")

    return {
        'train_acc': train_acc_history,
        'val_acc': val_acc_history,
        'best_val_acc': best_val_acc,
        'best_state': best_state
    }

# -----------------------
# Evaluate (confusion matrix & ROC) on a loader (uses mixed precision for forward)
# -----------------------
def evaluate_and_plot(model, loader, emotions):
    model.eval()
    all_true = []
    all_probs = []
    with torch.no_grad():
        for imgs, labels in loader:
            imgs = imgs.to(device, non_blocking=True)
            with autocast():
                outputs = model(imgs)
            probs = F.softmax(outputs, dim=1).cpu().numpy()
            all_probs.append(probs)
            all_true.append(labels.numpy())

    y_score = np.vstack(all_probs)
    y_true = np.concatenate(all_true)
    n_classes = len(emotions)
    y_pred = np.argmax(y_score, axis=1)

    cm = confusion_matrix(y_true, y_pred, labels=list(range(n_classes)))
    plot_confusion_matrix(cm, emotions)

    y_true_bin = label_binarize(y_true, classes=list(range(n_classes)))
    plot_multiclass_roc(y_true_bin, y_score, n_classes, emotions)

    print("\nClassification Report (Precision, Recall, F1-Score):")
    print(classification_report(y_true, y_pred, target_names=emotions, digits=4))

# -----------------------
# Helper to unnormalize a tensor for display
# -----------------------
def get_unnormalizer(mean, std):
    mean = np.array(mean)[:, None, None]
    std = np.array(std)[:, None, None]
    def unnormalize_to_image(tensor):
        t = tensor.clone().cpu().numpy()
        t = (t * std) + mean
        t = np.transpose(t, (1,2,0))
        return t
    return unnormalize_to_image

# -----------------------
# Predict helper (single image)
# -----------------------
def predict_single(model, img_path, transform, emotions):
    model.eval()
    img = Image.open(img_path).convert("RGB")
    x = transform(img).unsqueeze(0).to(device)
    with torch.no_grad():
        with autocast():
            out = model(x)
        probs = F.softmax(out, dim=1)
        idx = int(probs.argmax(dim=1).cpu().numpy()[0])
    return emotions[idx], float(probs.max().cpu().numpy())

# -----------------------
# MAIN
# -----------------------
def main():
    choice = input("Do you want to load the saved model? (yes/no): ").strip().lower()

    imagenet_mean = [0.485, 0.456, 0.406]
    imagenet_std  = [0.229, 0.224, 0.225]

    preprocess = transforms.Compose([
        transforms.Resize((IMG_SIZE, IMG_SIZE)),
        transforms.ToTensor(),
        transforms.Normalize(mean=imagenet_mean, std=imagenet_std),
    ])

    viz_transform = transforms.Compose([
        transforms.Resize((IMG_SIZE, IMG_SIZE)),
        transforms.ToTensor(),
    ])
    unnormalize_fn = get_unnormalizer(imagenet_mean, imagenet_std)

    print("Loading dataset paths...")
    image_paths, numeric_labels, emotions = load_multiple_datasets(ROOT_DATA_DIRS)
    n_classes = len(emotions)
    print(f"Number of labels (classes): {n_classes}")

    if n_classes == 0:
        raise RuntimeError("No emotion folders/images found. Check ROOT_DATA_DIRS paths.")

    try:
        os.makedirs(os.path.dirname(EMOTIONS_NPY), exist_ok=True)
    except Exception:
        pass
    np.save(EMOTIONS_NPY, emotions)

    X_train, X_val, y_train, y_val = train_test_split(
        image_paths, numeric_labels, test_size=0.3, random_state=SEED,
        stratify=numeric_labels if len(set(numeric_labels))>1 else None
    )

    train_ds = FaceEmotionDataset(X_train, y_train, transform=preprocess)
    val_ds   = FaceEmotionDataset(X_val, y_val, transform=preprocess)

    # DataLoaders: num_workers forced to 0, pin_memory enabled for faster host->device transfer
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True, drop_last=True)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

    # Visual check
    sample_path = X_train[0] if len(X_train) > 0 else (TEST_IMAGE if os.path.exists(TEST_IMAGE) else None)
    if sample_path:
        print("Displaying a sample before and after preprocessing...")
        img_orig = Image.open(sample_path).convert("RGB")
        img_after_t = preprocess(img_orig)
        img_after_vis = unnormalize_fn(img_after_t)
        fig, axs = plt.subplots(1, 2, figsize=(9,4))
        axs[0].imshow(img_orig)
        axs[0].set_title("Original")
        axs[0].axis("off")
        axs[1].imshow(np.clip(img_after_vis, 0, 1))
        axs[1].set_title("After Preprocessing (unnormalized for display)")
        axs[1].axis("off")
        plt.tight_layout()
        plt.show()
    else:
        print("No sample image found to visualize.")

    # Create model
    model = CNN_LSTM_Emotion(num_emotions=n_classes)

    if os.path.exists(DEFAULT_MODEL_PATH) and choice == "yes":
        print("Loading model from", DEFAULT_MODEL_PATH)
        state_dict = torch.load(DEFAULT_MODEL_PATH, map_location=device)
        if isinstance(state_dict, dict) and 'model_state' in state_dict:
            model.load_state_dict(state_dict['model_state'])
        else:
            model.load_state_dict(state_dict)
        model.to(device)
        evaluate_and_plot(model, val_loader, emotions)
        if os.path.exists(TEST_IMAGE):
            pred, conf = predict_single(model, TEST_IMAGE, preprocess, emotions)
            print(f"Single test image prediction: {pred} (conf {conf:.3f})")
        return

    model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=LR)
    scheduler = None  # optional

    print("Starting training...")
    stats = train_model(model, train_loader, val_loader, criterion, optimizer, epochs=EPOCHS, scheduler=scheduler)

    plot_accuracy(stats['train_acc'], stats['val_acc'])

    if stats['best_state'] is not None:
        model.load_state_dict(stats['best_state']['model_state'])
        print(f"Loaded best model from epoch {stats['best_state']['epoch']} with val acc {stats['best_state']['val_acc']:.2f}%")

    evaluate_and_plot(model, val_loader, emotions)

    np.save(EMOTIONS_NPY, emotions)
    final_save = os.path.splitext(DEFAULT_MODEL_PATH)[0] + "_final.pth"
    torch.save(model.state_dict(), final_save)
    print(f"Final model saved to {final_save}")
    print("[DONE]")

if __name__ == "__main__":
    main()
```
